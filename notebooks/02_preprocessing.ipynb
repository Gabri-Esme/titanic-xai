```python
import pandas as pd

# Load CSV
df = pd.read_csv("data/train.csv")

# Quick look at the data
print(df.head())
print(df.info())
```

```python
# Drop Irrelevant Colums

df = df.drop(columns=['PassengerId', 'Ticket'])
```

```python
# Handle missing values

# Age: fill missing with median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Embarked: fill missing with mode
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Cabin: create feature from first letter, fill missing with 'U'
df["CabinLetter"] = df["Cabin"].fillna("U").str[0]
df = df.drop(columns=['Cabin'])

# Extract title from Name
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
df = df.drop(columns=['Name'])
```

```python
# Encode categorical variables
df = pd.get_dummies(df, columns=["Sex", "Embarked", "CabinLetter", "Title"], drop_first=True)
```
